## Useful tricks and shortcuts

- **Line Magic (%), Cell magic (%%), and more** 
    1. %time, %%time, %timeit, %%timeit
    2. %matplotlib inline
    2. %%writefile filename.py
    3. %%run filename.py
    4. %store
    5. %pycat filename.py
    6. %load_ext 
    7. %autoreload 2
    8. ! pip install package (not suggested)
    9. ! ls
    10. ?
    
    
- **Outside Cell Shortcuts**
    0. Check shortcuts: H, CTRL SHIFT P
    1. Toggle line numbers: L
    2. Restart kernel: 0 0 
    2. Merge cells: SHIFT M
    3. Go back 1 step: Z 
    4. Markdown Cell: M
    5. Code Cell: Y
    6. Save: CTRL S
    7. Hide output of a cell: O
    8. Stop kernel :  I
    9. New cell: B
    10. Delete Cell: X
    
  
    
- **Inside Cell Shortcuts**
    1. Go back 1 step: CTRL Z
    2. Go Ahead 1 step: CTRL SHIFT Z
    3. Split cells: CTRL SHIFT -
    4. Comment a section: CTRL /
    5. Go Outisde: ESC
    6. Documentation in line: ALT + TAB    
    
    
- ** General **
    1. ALT
    2. TAB, SHIFT TAB  
    3. Print all outputs:
    ```  
        from IPython.core.interactiveshell import InteractiveShell
        InteractiveShell.ast_node_interactivity = "all"
    ```
    4. Suppress output: ;
